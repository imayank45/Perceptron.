```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv('/kaggle/input/sample-dataset-for-perceptron/linearly_separable_data.csv')
```

```python
data.head()
```

```python
sns.scatterplot(
    x='feature1', 
    y='feature2', 
    hue='output', 
    data=data
)
```

```python
X = data.iloc[:,0:2]
y = data.iloc[:,-1].astype(np.int_)
```

```python
p = Perceptron()
p.fit(X,y)
```

```python
# determite weights
p.coef_
```

```python
# determine bias
p.intercept_
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
plt.title('Perceptron Binary Classification')
```
